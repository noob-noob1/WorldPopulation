```python
df['Total_Population'] = df['Education_Post_Secondary'] + df['Education_Upper_Secondary'] +\
      df['Education_Lower_Secondary'] + df['Education_Primary'] + df['Education_Incomplete_Primary'] +\
          df['Education_No_Education'] + df['Population_Under_15']
```

Some of our columns seem to be skewed right or left. Let us investigate.

### 1. Density

```python
sns.histplot(df['Density'])
plt.ylabel('Frequency')
plt.xlabel('Number of People per Square Kilometer')
plt.xlim(0, 1000)
plt.show()
```

## It seem like higher populations are finding less and less area to live in.

```python
sns.scatterplot(data=df,y='Density',x='Year')
plt.ylabel('Number of People per Square Kilometer')
plt.show()
```

```python
sns.histplot(df['Education_Post_Secondary'])
plt.ylabel('Frequency')
plt.xlim(0,1000000)
plt.xlabel('Number of Educated People')
plt.show()
```

```python
sns.scatterplot(data=df,y='Education_Post_Secondary',x='Year')
plt.ylabel('Number of Educated People')
plt.show()
```

```python
sns.scatterplot(data=df,x='Education_Post_Secondary',y='Growth_Rate')
plt.ylabel('Growth Rate')
plt.show()
```

```python
sns.scatterplot(data=df,x='Education_Post_Secondary',y='Growth_Rate')
plt.xlim(1000000,10000000)
plt.xlabel('Number of Educated People')
plt.ylabel('Growth Rate')

plt.show()
```

```python
sns.scatterplot(data=df, x='Total_Population', y='Education_Post_Secondary', label='Educated Population')
sns.scatterplot(data=df, x='Total_Population', y='Education_No_Education', label='Not educated Population')
plt.legend()
plt.xlim(2000000000,10000000000)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
corr_matrix = df[['Year', 'Growth_Rate','Males_per_100_Females','Education_Post_Secondary']].corr()
sns.heatmap(corr_matrix,annot=True, cmap='coolwarm')
plt.xticks(rotation=45)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
corr_matrix = df[['Year', 'Education_Lower_Secondary','Education_Post_Secondary','Education_Upper_Secondary','Education_Primary']].corr()
sns.heatmap(corr_matrix,annot=True, cmap='coolwarm')
plt.xticks(rotation=45)
plt.show()
```
